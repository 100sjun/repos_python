```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터
categories = ['A', 'B', 'C', 'D']
data1 = [5, 7, 6, 8]  # 첫 번째 데이터 (누적)
data2 = [4, 6, 5, 7]  # 두 번째 데이터 (누적)
data3 = [3, 4, 4, 6]  # 세 번째 데이터 (누적)
additional_data = [10, 12, 11, 9]  # 독립적인 바 데이터

# 막대 폭 설정
bar_width = 0.4

# x축 위치 생성
x = np.arange(len(categories))

# 누적 바 그래프 생성
plt.bar(x, data1, width=bar_width, label='Data 1')
plt.bar(x, data2, width=bar_width, bottom=data1, label='Data 2')
plt.bar(x, data3, width=bar_width, bottom=np.array(data1) + np.array(data2), label='Data 3')

# 동일 x값에 별도 바 추가
plt.bar(x + bar_width, additional_data, width=bar_width, color='orange', label='Additional Data')

# x축 레이블 설정
plt.xticks(x + bar_width / 2, categories)

# 제목과 축 레이블 추가
plt.title('Stacked Bar + Additional Bar')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()

# 그래프 표시
plt.show()
```
