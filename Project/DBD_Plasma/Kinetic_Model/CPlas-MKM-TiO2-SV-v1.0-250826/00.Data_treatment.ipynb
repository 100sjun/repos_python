```python
df_mean['Power density [W/cm3]'] = df_mean['Discharge Power [W]'] / vol
df_mean['Residence Time [s]'] = vol / df_mean['CH4 flow [mL/min]'] * 60
df_mean
```

```python
df_mean['Power density [W/cm3]'].iloc[0:4].mean()
df_mean.to_csv('kinetic_data.csv', index=False)
```

```python
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
axs = axs.flatten()

# voltage가 15kV인 데이터만 필터링
df_filtered = df_mean[df_mean['Applied Voltage (kV)'] == 15.0]

components = ['mol%H2', 'mol%CH4', 'mol%C2H6', 'mol%C2H4', 'mol%C2H2', 'mol%C3H8', 'mol%C3H6', 'mol%C4H10', 'mol%C5H12']
for i, component in enumerate(components):
    axs[i].scatter(df_filtered['CH4 flow [mL/min]'], df_filtered[component], marker='o')
    axs[i].set_title(f'{component} vs CH4 flow')
    axs[i].set_xlabel('CH4 flow [mL/min]')
    axs[i].set_ylabel(component)

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))
axs = axs.flatten()

# CH4 flow가 30인 데이터만 필터링
df_filtered_ch4_30 = df_mean[df_mean['CH4 flow [mL/min]'] == 30]

# voltage 별로 데이터 필터링
for voltage in df_filtered_ch4_30['Applied Voltage (kV)'].unique():
    df_filtered_voltage = df_filtered_ch4_30[df_filtered_ch4_30['Applied Voltage (kV)'] == voltage]
    for i, component in enumerate(components):
        axs[i].scatter([voltage] * len(df_filtered_voltage), df_filtered_voltage[component], label=f'Voltage: {voltage}', marker='o')

# 각 subplot에 제목과 레이블 추가
for i, component in enumerate(components):
    axs[i].set_title(f'{component} vs Voltage')
    axs[i].set_xlabel('Applied Voltage (kV)')
    axs[i].set_ylabel(component)
    axs[i].legend()

plt.tight_layout()
plt.show()
```
