```python
import os 
import re
```

```python
def natural_sort_key(filename):
    """Extract numbers from filename for natural sorting."""
    return [int(text) if text.isdigit() else text.lower() for text in re.split(r'(\d+)', filename)]
```

```python
folder_path = './crossA_DB'
```

```python
file_list = [f for f in os.listdir(folder_path) if f.endswith(".txt")]
```

```python
file_list.sort(key=natural_sort_key)
```

```python
file_list
```

```python
for index, filename in enumerate(file_list):
    old_path = os.path.join(folder_path, filename)  # 기존 파일 경로
    new_filename = f"{index}.txt"  # 새로운 파일 이름
    new_path = os.path.join(folder_path, new_filename)  # 새로운 파일 경로
    print(old_path,new_path)

    # 파일 이름 변경
    os.rename(old_path, new_path)
```
