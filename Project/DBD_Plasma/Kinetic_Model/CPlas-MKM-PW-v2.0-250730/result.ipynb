```python
species = []
with open('qt_species_list.txt','r') as f:
    for line in f:
        comp = line[2:]
        species.append(comp.strip())
    f.close()
```

```python
df_sp = pd.read_csv('qt_densities.txt', sep=r'\s+', header=0, names=['Time [s]']+species)
H2 = (df_sp['H2'])
CH4 = (df_sp['CH4'] + df_sp['CH4(V13)'] + df_sp['CH4(V24)'])
C2H2 = (df_sp['C2H2'] + df_sp['C2H2(V2)'] + df_sp['C2H2(V5)'] + df_sp['C2H2(V13)'])
C2H4 = (df_sp['C2H4'] + df_sp['C2H4(V1)'] + df_sp['C2H4(V2)'])
C2H6 = (df_sp['C2H6'] + df_sp['C2H6(V13)'] + df_sp['C2H6(V24)'])
C3H6 = (df_sp['C3H6'] + df_sp['C3H6(V)'])
C3H8 = (df_sp['C3H8'] + df_sp['C3H8(V1)'] + df_sp['C3H8(V2)'])
C4H10 = (df_sp['C4H9H'])
C5H12 = (df_sp['C5H12'])
C = (df_sp['C'])
CH = (df_sp['CH'])
CH2 = (df_sp['CH2'])
CH3 = (df_sp['CH3'])
C2H3 = (df_sp['C2H3'])
C2H5 = (df_sp['C2H5'])
C3H7 = (df_sp['C3H7'])
H = (df_sp['H'])
CH3_plus = (df_sp['CH3^+'])
CH4_plus = df_sp['CH4^+']
CH5_plus = df_sp['CH5^+']
C2H2_plus = df_sp['C2H2^+']
C2H4_plus = df_sp['C2H4^+']
C2H5_plus = df_sp['C2H5^+']
C2H6_plus = df_sp['C2H6^+']
C3H6_plus = df_sp['C3H6^+']
C3H8_plus = df_sp['C3H8^+']
```

```python
rH2 = H2 - 0.5*CH3 - 0.5*CH3_plus - CH2 - 0.5*C2H3 + 0.5*C2H5 + 0.5*C2H5_plus + 0.5*H + 0.5*C3H7 + 0.5*CH5_plus - 1.5*CH
rCH4 = CH4 + CH3 + CH2 + CH + CH5_plus + CH3_plus + CH4_plus
rC2H6 = C2H6 + C2H6_plus
rC2H4 = C2H4 + C2H4_plus + C2H3 + C2H5 + C2H5_plus
rC2H2 = C2H2 + C2H2_plus
rC3H8 = C3H8 + C3H8_plus
rC3H6 = C3H6 + C3H6_plus + C3H7
rC4H10 = C4H10
rC5H12 = C5H12
rC = C
rtot = rCH4 + rC2H6 + rC2H4 + rC2H2 + rC3H8 + rC3H6 + rC4H10 + rC5H12 + rC + rH2
```

```python
time = df_sp['Time [s]']
X_CH4 = (rCH4[0] - rCH4)/rCH4[0]*100
S_H2 = (rH2)/(rCH4[0] - rCH4)*100*0.5
S_C2H6 = (rC2H6)/(rCH4[0] - rCH4)*100*2
S_C2H4 = (rC2H4)/(rCH4[0] - rCH4)*100*2
S_C2H2 = (rC2H2)/(rCH4[0] - rCH4)*100*2
S_C3H8 = (rC3H8)/(rCH4[0] - rCH4)*100*3
S_C3H6 = (rC3H6)/(rCH4[0] - rCH4)*100*3
S_C4H10 = (rC4H10)/(rCH4[0] - rCH4)*100*4
S_C5H12 = (rC5H12)/(rCH4[0] - rCH4)*100
S_C = (rC)/(rCH4[0] - rCH4)*100
```

```python
rx = []
with open('qt_reactions_list.txt','r') as f:
    for line in f:
        comp = line[3:]
        rx.append(comp.strip())
    f.close()
```

```python
df_rx = pd.read_csv('qt_rates.txt', sep=r'\s+', header=0, names=['Time [s]']+rx)
```

```python
reaction_extents = {}
reaction_extents['Time [s]'] = df_rx['Time [s]']
for i in range(len(rx)):
    reaction = []
    for j in range(len(df_rx)):
        if j == 0:
            reaction.append(0)
        else:
            reaction.append(reaction[j-1] +df_rx[rx[i]].iloc[j-1]*(df_rx['Time [s]'].iloc[j]-df_rx['Time [s]'].iloc[j-1]))
    reaction_extents[rx[i]] = reaction
```

```python
df_rex = pd.DataFrame(reaction_extents)
```
