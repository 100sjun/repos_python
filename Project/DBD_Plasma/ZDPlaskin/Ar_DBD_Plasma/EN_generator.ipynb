```python
import numpy as np
import pandas as pd
from scipy import constants
import matplotlib.pyplot as plt
```

```python
# 유효숫자 2개로 반올림하는 함수
def round_to_2_sig_figs(x):
    if pd.isna(x) or x == 0:  # NaN 값이나 0 처리
        return x
    if x < 1e-10:
        return 0
    return float(f'{x:.5g}')
```

```python
kb = constants.Boltzmann
me = constants.electron_mass # electron mass (kg)
qe = constants.elementary_charge # electron charge (C)
print(np.format_float_scientific((me),precision=2))
```

```python
t = np.arange(0,1e-4,5e-7)
V0 = 30
P = 101325
T = 400
N = P/kb/T
d = 1e-4
f = 50000
E = abs(-V0*np.sin(2*50000*np.pi*t)/d)
EN = E/N*1e21
```

```python
# electron flux
N
```

```python
plt.plot(t,EN)
plt.show()
```

```python
df = pd.DataFrame({'Time_s':t, 'Field_Td': EN})
df_rounded = df.applymap(round_to_2_sig_figs)
df_rounded.to_csv('data_EN.dat',sep='\t', index=False)
```
