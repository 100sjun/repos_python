```python
import pandas as pd
import re
```

```python
def correct_chemical_formula(formula):
    """
    주어진 화학식을 검사하고 필요한 경우 수정합니다.
    공백을 기준으로 화학종을 나누고, 화학종 사이에 '+' 또는 '->'가 없으면 '+'를 추가합니다.
    """
    parts = formula.split()
    corrected = []
    
    for i, part in enumerate(parts):
        if i > 0 and part not in ['+', '->'] and parts[i-1] not in ['+', '->']:
            corrected.append('+')
        corrected.append(part)
    
    return ' '.join(corrected)
```

```python
file_path = './ZDPlaskin/kinetic_DB/'
i_list = list(range(1,247))

for i in i_list:
    file_name = 'reaction_%d.csv'%i
    df = pd.read_csv(file_path+file_name)
    df['Formula'] = df['Formula'].apply(correct_chemical_formula)

    if df['Type'].loc[0] == 'IONIZATION':
        corr_reaction = df['Reaction'].loc[0]
        corr_reaction = corr_reaction[:-1] + "^+"
        df['Reaction'] = corr_reaction

        corr_formula = df['Formula'].loc[0]
        corr_formula = corr_formula[:-1] + "^+"
        df['Formula'] = corr_formula

    df.to_csv(file_path+file_name)

```
