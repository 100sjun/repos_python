```python
import matplotlib.pyplot as plt

# Reaction data
reactions = [
    ("CH4 -> CH3 H", 8.8, 1),
    ("CH4 -> CH2 H2", 9.4, 1),
    ("CH4 -> CH H2 H", 12.5, 1),
    ("CH4 -> C H2 H2", 14, 1),
    ("C2H6 -> C2H5 H", 7.45, 2),
    ("C2H6 -> C2H4 H", 4, 2),
    ("C2H6 -> C2H3 H2 H", 9.4, 2),
    ("C2H6 -> C2H2 H2 H2", 6.2, 2),
    ("C2H4 -> C2H3 H", 6.9, 2),
    ("C2H4 -> C2H2 H2", 5.8, 2),
    ("C2H4 -> C2H2 H H", 6.5, 2),
    ("C2H4 -> CH3 CH", 8.7, 2),
    ("C2H4 -> CH2 CH2", 8.9, 2),
    ("C2H2 -> CH CH", 10.6, 2),
    ("C3H8 -> C3H7 H", 5.82, 3),
    ("C3H8 -> C3H6 H2", 2.66, 3),
    ("C3H8 -> C3H4 H2 H2", 4.26, 3),
    ("C3H8 -> C2H6 CH2", 5.7, 3),
    ("C3H8 -> C2H5 CH3", 5.05, 3),
    ("C3H8 -> C2H4 CH4", 2.1, 3),
    ("C3H6 -> C3H4 H2", 3.76, 3),
    ("C3H6 -> C2H4 CH2", 5.82, 3),
    ("C3H6 -> C2H3 CH3", 5.81, 3),
    ("C3H6 -> C2H2 CH4", 3.35, 3),
]

# Separate data by number of carbon atoms (C)
C1_reactions = [(reaction[1], reaction[0]) for reaction in reactions if reaction[2] == 1]
C2_reactions = [(reaction[1], reaction[0]) for reaction in reactions if reaction[2] == 2]
C3_reactions = [(reaction[1], reaction[0]) for reaction in reactions if reaction[2] == 3]

# Plotting the scatter plot
plt.figure(figsize=(10, 6))

# Plotting C1 reactions
C1_x = range(len(C1_reactions))
C1_y = [r[0] for r in C1_reactions]
C1_labels = [r[1] for r in C1_reactions]
plt.plot(C1_x, C1_y, 'bo-', label='\n'.join(C1_labels))  # Line for C1 with all reactions in label

# Plotting C2 reactions
C2_x = range(len(C1_reactions), len(C1_reactions) + len(C2_reactions))
C2_y = [r[0] for r in C2_reactions]
C2_labels = [r[1] for r in C2_reactions]
plt.plot(C2_x, C2_y, 'ro-', label='\n'.join(C2_labels))  # Line for C2 with all reactions in label

# Plotting C3 reactions
C3_x = range(len(C1_reactions) + len(C2_reactions), len(C1_reactions) + len(C2_reactions) + len(C3_reactions))
C3_y = [r[0] for r in C3_reactions]
C3_labels = [r[1] for r in C3_reactions]
plt.plot(C3_x, C3_y, 'go-', label='\n'.join(C3_labels))  # Line for C3 with all reactions in label

# Adding title and labels
plt.title('Threshold Energies of Dissociative excitation')
plt.xlabel('Reaction Index')
plt.ylabel('Threshold Energy (kcal/mol)')

# Move legend to the right side of the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='small', title='Reactions by Group')

# Show plot
plt.tight_layout()
plt.show()

```
