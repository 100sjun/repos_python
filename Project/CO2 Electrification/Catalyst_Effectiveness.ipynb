```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 2차 미분 방정식: y'' + 2y' + y = 0
# 이를 1차 연립방정식으로 변환:
# let y1 = y, y2 = y'
# y1' = y2
# y2' = -2y2 - y1

def ode_system(t, y):
    # 확산과 반응만 포함하는 2차 미분방정식: D*y'' - k*y = 0
    # y1 = y, y2 = y'
    # y1' = y2
    # y2' = (k/D)*y1

    D = 1.0  # 확산계수 (예시값)
    k = 1.0  # 반응속도상수 (예시값)

    y1, y2 = y
    dy1dt = y2
    dy2dt = (k/D)*y1
    return [dy1dt, dy2dt]

# 초기 조건: y(0) = 1, y'(0) = 0
y0 = [1, 0]

# 적분 구간
t_span = (0, 10)
t_eval = np.linspace(t_span[0], t_span[1], 200)

sol = solve_ivp(ode_system, t_span, y0, t_eval=t_eval)

plt.plot(sol.t, sol.y[0], label='y(t)')
plt.plot(sol.t, sol.y[1], label="y'(t)")
plt.xlabel('t')
plt.ylabel('값')
plt.legend()
plt.title('2차 미분방정식 해 (solve_ivp)')
plt.show()
```
