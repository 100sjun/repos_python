```python
fig, axs = plt.subplots(8, 1, figsize=(16, 64))
fig.suptitle('  plot ', fontsize=20)

#    
explist = sorted(df['exp'].unique())
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan',
          'navy', 'gold', 'teal', 'coral', 'lime', 'indigo', 'magenta', 'black', 'peru', 'aqua']

# 8x1 subplot    
plot_cols = ['VF', 'VA', 'VB', 'CFLA', 'CALA', 'CFK', 'CBK', 'I']

for idx, col in enumerate(plot_cols):
    for i, exp in enumerate(explist):
        axs[idx].plot(
            df[df['exp'] == exp]['t'],
            df[df['exp'] == exp][col],
            color=colors[i % len(colors)],
            label=f'exp {exp}'
        )
        #        
        exp_data = df[df['exp'] == exp]
        if not exp_data.empty:
            axs[idx].text(
                exp_data['t'].values[-1], 
                exp_data[col].values[-1], 
                str(exp), 
                color=colors[i % len(colors)],
                fontsize=10,
                fontweight='bold'
            )
    axs[idx].set_title(col)
    axs[idx].legend(fontsize=8, loc='best', ncol=2)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
