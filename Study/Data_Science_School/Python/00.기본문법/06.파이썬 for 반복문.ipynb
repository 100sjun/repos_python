```python
for i in range(20):
    if i < 10:
        print('*'*i)
    else:
        print((20-i)*'*')
```

```python
for i in range(11):
    if i <= 5:
        print(' '*(5-i) + '*'*(1+2*i))
    else:
        print(' '*(i - 5) + '*'*(1+2*(10-i)))
```

# 중첩 for 반복문

```python
for i in range(4):
    for j in range(4):
        print(i + j, end = ' ')
    print()
```

연습 문제 2.7.5

```python
a = 1024

for i in range(2):
    if i == 0:
        b = a*2
    else:
        b = a/2
    for j in range(2):
        if j == 0:
            c = b*2
        else:
            c = b/2
        for k in range(2):
            if k == 0:
                d = c*2
                print(d)
            else:
                d = c/2
                print(d)
```

연습 문제 2.7.6

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            s = (i+1)**3+(j+1)**3
            d = (k+1)**3
            print(f'left: {s}, right: {d}, so {s==d}')
```

# 카운터 변수가 변하는 경우

반복문에서 가장 복잡한 형태는 안쪽 반복문의 반복 횟수가 바깥쪽 반복문의 카운터 변수에 따라 변하는 상황이다.

```python
for j in range(10):
    sum = 0
    for i in range(j+1):
        sum = sum + (i + 1)
    print(sum)
```
