```python
import jax.numpy as jnp
from jax import grad
from jax import jit
```

```python
# 기본 연산

## 배열 생성
x = jnp.array([1., 2., 3.])
y = jnp.array([0.1, 0.2, 0.3])

## 기본 연산
sum_xy = jnp.add(x, y)
print(f'Sum: {sum_xy}')
```

```python
# 자동 미분

## 손실 함수 정의
def loss_fn(w):
    return jnp.sum(w ** 2)

## 기울기 계산
grad_loss = grad(loss_fn)
print(f'Gradient: {grad_loss(x)}')
```

```python
# JIT 컴파일

## JIT을 이용한 함수
@jit
def multiply(a, b):
    return a * b

print(f'JIT Multiply: {multiply(x, y)}')
```
